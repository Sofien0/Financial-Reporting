```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# ⚙️ Configuration du navigateur Chrome (avec affichage)
options = webdriver.ChromeOptions()
# ➡️ Ne pas utiliser le mode headless pour voir le navigateur
# options.add_argument("--headless")  # Commenté pour affichage

# 🚀 Lancer le navigateur
driver = webdriver.Chrome(options=options)

# 🌐 Aller à la page SASB
url = "https://sasb.ifrs.org/company-use/sasb-reporters/"
print(f"📡 Accès à la page : {url}")
driver.get(url)
time.sleep(5)  # Laisser le temps de charger les données

# 📥 Récupérer les lignes du tableau
rows = driver.find_elements(By.CSS_SELECTOR, "div[data-testid='reports-search-results'] table tbody tr")
print(f"🔍 Nombre de lignes trouvées : {len(rows)}")

data = []

# 🔁 Parcours ligne par ligne
for i, row in enumerate(rows):
    cols = row.find_elements(By.TAG_NAME, "td")
    if len(cols) >= 6:
        company_name = cols[0].text.strip()
        industry = cols[1].text.strip()
        sector = cols[2].text.strip()
        country = cols[3].text.strip()
        report_type = cols[4].text.strip()
        report_year = cols[5].text.strip()

        # 🔗 Récupération du lien
        try:
            report_link = cols[0].find_element(By.TAG_NAME, "a").get_attribute("href")
            source = "SASB"
        except:
            report_link = ""
            source = ""

        # 🖨️ Afficher dans la console
        print(f"\n#{i+1} 🔹 {company_name}")
        print(f"    📦 Industry  : {industry}")
        print(f"    🏭 Sector    : {sector}")
        print(f"    🌍 Country   : {country}")
        print(f"    📑 Type      : {report_type}")
        print(f"    📆 Year      : {report_year}")
        print(f"    🔗 Report    : {report_link if report_link else '❌ Aucun lien'}")
        print(f"    🏷️ Source    : {source if source else '—'}")

        # 📤 Ajouter aux données
        data.append({
            "Company Name": company_name,
            "Industry": industry,
            "Sector": sector,
            "Country": country,
            "Report Type": report_type,
            "Report Year": report_year,
            "Report Link": report_link,
            "Source": source
        })

# 🧾 Convertir en DataFrame
df = pd.DataFrame(data)

# 💾 Exporter
df.to_csv("sasb_reports_selenium.csv", index=False, encoding="utf-8-sig")
print("\n✅ Fichier 'sasb_reports_selenium.csv' enregistré avec succès.")

# 🧹 Fermer le navigateur
driver.quit()
```
