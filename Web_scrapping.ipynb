```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# âš™ï¸ Configuration du navigateur Chrome (avec affichage)
options = webdriver.ChromeOptions()
# â¡ï¸ Ne pas utiliser le mode headless pour voir le navigateur
# options.add_argument("--headless")  # CommentÃ© pour affichage

# ğŸš€ Lancer le navigateur
driver = webdriver.Chrome(options=options)

# ğŸŒ Aller Ã  la page SASB
url = "https://sasb.ifrs.org/company-use/sasb-reporters/"
print(f"ğŸ“¡ AccÃ¨s Ã  la page : {url}")
driver.get(url)
time.sleep(5)  # Laisser le temps de charger les donnÃ©es

# ğŸ“¥ RÃ©cupÃ©rer les lignes du tableau
rows = driver.find_elements(By.CSS_SELECTOR, "div[data-testid='reports-search-results'] table tbody tr")
print(f"ğŸ” Nombre de lignes trouvÃ©es : {len(rows)}")

data = []

# ğŸ” Parcours ligne par ligne
for i, row in enumerate(rows):
    cols = row.find_elements(By.TAG_NAME, "td")
    if len(cols) >= 6:
        company_name = cols[0].text.strip()
        industry = cols[1].text.strip()
        sector = cols[2].text.strip()
        country = cols[3].text.strip()
        report_type = cols[4].text.strip()
        report_year = cols[5].text.strip()

        # ğŸ”— RÃ©cupÃ©ration du lien
        try:
            report_link = cols[0].find_element(By.TAG_NAME, "a").get_attribute("href")
            source = "SASB"
        except:
            report_link = ""
            source = ""

        # ğŸ–¨ï¸ Afficher dans la console
        print(f"\n#{i+1} ğŸ”¹ {company_name}")
        print(f"    ğŸ“¦ Industry  : {industry}")
        print(f"    ğŸ­ Sector    : {sector}")
        print(f"    ğŸŒ Country   : {country}")
        print(f"    ğŸ“‘ Type      : {report_type}")
        print(f"    ğŸ“† Year      : {report_year}")
        print(f"    ğŸ”— Report    : {report_link if report_link else 'âŒ Aucun lien'}")
        print(f"    ğŸ·ï¸ Source    : {source if source else 'â€”'}")

        # ğŸ“¤ Ajouter aux donnÃ©es
        data.append({
            "Company Name": company_name,
            "Industry": industry,
            "Sector": sector,
            "Country": country,
            "Report Type": report_type,
            "Report Year": report_year,
            "Report Link": report_link,
            "Source": source
        })

# ğŸ§¾ Convertir en DataFrame
df = pd.DataFrame(data)

# ğŸ’¾ Exporter
df.to_csv("sasb_reports_selenium.csv", index=False, encoding="utf-8-sig")
print("\nâœ… Fichier 'sasb_reports_selenium.csv' enregistrÃ© avec succÃ¨s.")

# ğŸ§¹ Fermer le navigateur
driver.quit()
```
